```python
X = np.arange(1,13)
fig = plt.figure(figsize=(8, 3))

ax = fig.add_axes([0,0,1,1])
plt.title('Curah Hujan Setiap Tahun',weight='bold',fontsize='15')
ax.set_xlabel('Bulan',weight='bold',fontsize='13')
ax.set_ylabel('Curah Hujan (mm)', weight='bold',fontsize='13')
ax.bar(X + 0.00, data['1981-1990'], color = '#00bfff', width = 0.20)
ax.bar(X + 0.20, data['1991-2000'], color = '#ff4d4d', width = 0.20)
ax.bar(X + 0.40, data['2001-2010'], color = '#00b359', width = 0.20)
ax.legend(['1981-1990', '1991-2000', '2001-2010'], ncol=3, loc='upper center',fontsize='12', frameon=False)

for d in [100, 200, 300, 400, 500, 600, 700]:
  if d != 700:
    plt.axhline(y=d, color='black', linestyle='-', linewidth=0.15)
  else:
    plt.axhline(y=d, color='black', linestyle=' ', linewidth=0.15)

ax.set_xticks(X + 0.20)
ax.set_xticklabels(np.arange(1, 13))

plt.show()
```

```python
data1 = [23, 25.2, 25.5, 24.8, 22.5, 22.2, 23, 22, 25.9]
data2 = [23.5, 25.3, 25.1, 25.5, 24.9, 25, 24.5, 24, 26]
data3 = [24, 25, 23.8, 23.9, 24, 24.1, 24, 22.9, 23]
dates = pd.date_range(start='2013-07-24', periods=9, freq='3H')

plt.figure(figsize=(10, 4))

plt.plot(dates, data1, marker='o', color='#00bfff', label='Asimilasi')
plt.plot(dates, data3, marker='o', color='#e6e600', label='Observasi')
plt.plot(dates, data2, marker='o', color='#ff4d4d', label='Non Asimilasi')

plt.xlabel('Waktu', weight='bold',fontsize='13')
plt.ylabel('Celcius', weight='bold',fontsize='13')
plt.title('Suhu Permukaan',weight='bold',fontsize='15')
plt.legend(loc="lower right")

for d in [20,21, 22, 23, 24, 25, 26, 27]:
  plt.axhline(y=d, color='gray', linestyle='-', linewidth=0.3)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M UTC'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())

plt.show()

# Dapatkan tanggal
now = datetime.utcnow()
print("Tanggal saat ini (UTC):", now.date())
```
